```python
from classy import Class
my_model = Class()

# It's possible to use sigma8 to set the value at z=0
settings = {
    'output': 'mPk',
    'sigma8': 0.8,
}

my_model.set(settings)
my_model.compute()  
print(f'sigma8 = {my_model.sigma8()}')

```

```python
my_model = Class()

#Alternatively, you can input S8 = sigma8 * (Omega_b + Omega_cdm/0.3)**0.5
settings = {
    'output': 'mPk',
    'S8': 0.8,
}

my_model.set(settings)
my_model.compute()  

print(f'sigma8 = {my_model.sigma8()}')

```

##### The value of sigma8 is calculated for each input value and is equal to the input value with a very low uncertainty

```python
import numpy as np  
import matplotlib.pyplot as plt

input_sigma8s = np.linspace(0.2, 1.5, 100)
calculated_sigma8s = []
for sigma in input_sigma8s:
    my_model = Class()
    my_model.set({'output': 'mPk','sigma8_input': sigma})
    my_model.compute()
    calculated_sigma8s.append(my_model.sigma8())
```

```python
fig, ax = plt.subplots()
sigma8_diffs = [(input - calculated_sigma8)/input for input, calculated_sigma8 in zip(input_sigma8s, calculated_sigma8s)]
ax.set(xlabel=r'$\sigma_8$ input', ylabel='Relative difference', title=r'Relative difference between input and calculated $\sigma_8$')
ax.plot(input_sigma8s, sigma8_diffs, '-')
```

The computation time of the $\sigma_8$-routine has been lower considerably. 
The following graphs show that the mean computation time is now ~7.6 times lower and there is no clear dependency on the input value.
